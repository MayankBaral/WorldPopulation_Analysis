```python
def log_transform(data,col):
    for colname in col:
        if(data[colname]==1.0).all():
            data[colname+'_log']=np.log(data[colname]+1)
        else:
            data[colname+'_log']=np.log(data[colname])
```

```python
log_transform(new_dfc,['population','pop_density'])
```

```python
for col in new_dfc[['population_log','pop_density_log']]:
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    sns.histplot(new_dfc[col],kde=True)
    plt.subplot(1,2,2)
    sns.boxplot(x=new_dfc[col])
    plt.show()
```

### As we can see net migration and migration percentage have many missing values and they have right skewness, so before handling skewness we need to handle missing values

#### To handle the missing data I'll use prev. year and current year population

```python
new_dfc.sort_values(by=['country','year'],inplace=True)
for col in new_dfc['country'].unique():
    country_mask = new_dfc['country'] == col
    missing_mask = new_dfc['net_migration'].isnull()  
    new_dfc.loc[country_mask & missing_mask,'net_migration']=new_dfc.loc[country_mask & missing_mask,'population'].diff()
    
new_dfc.head(60)
```

```python
new_dfc.dropna(subset=['net_migration'],inplace=True)
```

```python
for col in new_dfc['migration_perc']:
    miss_mask = new_dfc['migration_perc'].isnull()
    new_dfc.loc[miss_mask,'migration_perc'] = new_dfc.loc[miss_mask, 'net_migration'] / new_dfc.loc[miss_mask, 'population']

new_dfc.head(60)
```

```python
for col in new_dfc[['net_migration','migration_perc']]:
    plt.figure(figsize=(20,10))
    plt.subplot(1,2,1)
    sns.histplot(new_dfc[col],kde=True)
    plt.subplot(1,2,2)
    sns.boxplot(x=new_dfc[col])
    plt.show()
```

```python
fig, axes = plt.subplots(3,1, figsize=(20,40))
sns.countplot(ax=axes[0], x='region', data=new_dfc, order=new_dfc['region'].value_counts().index)
sns.countplot(ax=axes[1], x='incomeLevel', data=new_dfc, order=new_dfc['incomeLevel'].value_counts().index)
sns.countplot(ax=axes[2], x='adminregion', data=new_dfc, order=new_dfc['adminregion'].value_counts().index)
axes[0].tick_params(labelrotation=45)
axes[1].tick_params(labelrotation=45)
axes[2].tick_params(labelrotation=45);
```
